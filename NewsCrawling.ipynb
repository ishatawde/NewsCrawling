```python
#dynamic url, date-time format in the format that the URL accepts
from datetime import date
today=date.today()

d= today.strftime("%m-%d-%y")
print("date=",d)


#defining it in an url object, substitute d 
cnn_url="https://www.cnn.com/world/live-news/coronavirus-pandemic-{}-intl/index.html".format(d)


print(cnn_url)


from bs4 import BeautifulSoup
import requests # create the url


#get the entire html text
html = requests.get(cnn_url).text

#assign it to a text object
#calling BS function and passing html, create a soup object & then I can go and extract specific context
soup = BeautifulSoup(html)
print(soup.title)
```

```python
# who all are involved- countires, known entitites
import spacy
spacy.prefer_gpu()
nlp = spacy.load('en_core_web_sm')

nbc_url = 'https://www.nbcnews.com/health/coronavirus'

urls = [cnn_url, nbc_url]
formats=['html.parser', 'html.parser']
tags=['h2','h2']
website=['CNN', 'NBC']


crawl_len = 0
for url in urls:
    print("Crawing web page ........{}".format(url))
    response = requests.get(url)
    soup= BeautifulSoup(response.content, formats[crawl_len])
    
    for link in soup.find_all(tags[crawl_len]):
        
        if(len(link.text.split(" "))> 4):
            print("Headline : {}".format(link.text))
            
            entities=[]
            for ent in nlp(link.text).ents:
                print('\tText : {}, Entity : {}'.format(ent.text, ent.label_))
                
    crawl_len=crawl_len+1
```

```python
crawl_len = 0
news_dict=[]
for url in urls:
    
    response = requests.get(url)
    soup= BeautifulSoup(response.content, formats[crawl_len])
    
    for link in soup.find_all(tags[crawl_len]):
        #print(link.text)
        if(len(link.text.split(" "))> 4):
            
            entities=[]
            entities=[(ent.text, ent.label_) for ent in nlp(link.text).ents]
            news_dict.append({'website':website[crawl_len], 'url': url, 'headline': link.text, 'entities': entities})

                
    crawl_len=crawl_len+1
```

```python
import pandas as pd
pd.set_option("display.max_rows", None, "display.max_columns", None)

news_df=pd.DataFrame(news_dict)

pd.set_option('max_colwidth', 800)

news_df.head(20)

pd.concat([news_df[['website', 'url', 'headline']], news_df['entities'].apply(pd.Series)], axis=1)
```

```python
#Classification for Online News Popularity
#Importing the Libraries
import pandas as pd
#Importing the dataset
dataset = pd.read_csv('OnlineNewsPopularity.csv')
dataset.head()
```

```python
X = dataset.iloc[:,2:-2].values
Y = dataset[' shares']
#Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(random_state=42)
rf.fit(X_train, Y_train)
```

```python
rf_df = pd.DataFrame(rf.predict(X_train),list(Y_train))
```

```python
rf_df.reset_index(level=0, inplace=True)
final_df = rf_df.rename(index=str, columns={"index": "Actual shares in data", 0: "Predicted shares in data"})
final_df.head(50)
```

```python
#Using gradient Boost Classfier
# Have classified the data into classes - above 1400 (4) is popular and below 1400 in unpopular (2)
dataset['Class'] = [4 if x>1400 else 2 for x in dataset[' shares']]
#dataset.isnull()
dataset.head(10)
```

```python
X = dataset.iloc[:,2:-2].values
Y = dataset.iloc[:, -1].values
#Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
#Fitting Gradient Boosting to Training set
from sklearn.ensemble import GradientBoostingClassifier
classifier = GradientBoostingClassifier()
classifier.fit(X_train, Y_train)
#Predicting the Test set results
Y_pred = classifier.predict(X_test)
#Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, Y_pred)
cm
```

```python
gf_res = pd.DataFrame(classifier.predict(X_train),list(Y_train))
gf_res.reset_index(level=0, inplace=True)
gf_res_renamed = gf_res.rename(index=str, columns={"index": "Actual Class", 0: "Predicted Class"})
gf_res_renamed.head(50)
```
